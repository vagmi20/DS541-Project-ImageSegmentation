```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import cv2 
import os 
from random import shuffle 
from tqdm import tqdm 
from PIL import Image
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
import shutil

import warnings
warnings.filterwarnings('ignore')
import os
```

```python
def find_last_image_per_patient(filePath):
    counter = 0
    patient_num = ""
    image_num = ""
    last_image_per_patient = []
    with open(filePath, 'r') as file:
        for line in file:
            if(counter == 0):
                patient_num = line[0:8]
                image_num = line[9:-4]
                counter+=1
                continue 
            if patient_num != line[0:8]:
                image = patient_num + "_" + image_num + "png"
                last_image_per_patient.append(image)
                patient_num = line[0:8]
                image_num = line[9:-4]
            else:
                image_num = line[9:-4]
                continue
        
        image = patient_num + "_" + image_num + ".png" # last image
        last_image_per_patient.append(image)
        file.close()

    return last_image_per_patient
```

```python
last_image_per_patient_train_val = find_last_image_per_patient("../data/CXR8/train_val_list.txt")
last_image_per_patient_test = find_last_image_per_patient("../data/CXR8/test_list.txt")
image_subset = sorted(last_image_per_patient_test + last_image_per_patient_train_val)
```

```python
image_dir = "../data/CXR8/images"
counter = 0
dir = 1
for image in image_subset:
    if not os.path.isfile(f"{image_dir}/images_{dir}/{image}"):
        dir += 1
    path = os.path.join(image_dir, f"images_{dir}", image)
    print(f"Copying {image}...")
    shutil.copy(path, f"{image_dir}/image_subset")
    counter+=1
        
print(f"File Count: {counter}")
```

```python
image_dir = "../data/CXR8/images"
counter = 0
dir = 1
for image in last_image_per_patient_train_val:
    if not os.path.isfile(f"{image_dir}/images_{dir}/{image}"):
        dir += 1
    path = os.path.join(image_dir, f"images_{dir}", image)
    print(f"Copying {image}...")
    shutil.copy(path, f"{image_dir}/train_val_images")
    counter+=1
        
print(f"File Count: {counter}")
```

```python
image_dir = "../data/CXR8/images"
counter = 0
dir = 1
for image in last_image_per_patient_test:
    if not os.path.isfile(f"{image_dir}/images_{dir}/{image}"):
        dir += 1
    path = os.path.join(image_dir, f"images_{dir}", image)
    print(f"Copying {image}...")
    shutil.copy(path, f"{image_dir}/test_images")
    counter+=1
        
print(f"File Count: {counter}")
```
